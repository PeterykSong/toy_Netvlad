# Positive / Negative 셋트를 구성하기 위한 거리 계산

```python
import math
from scipy.io import loadmat
import collections

mat = loadmat("./data/Pittsburgh250k/groundtruth/groundtruth/pittsburgh_database_10586_utm.mat")#,squeeze_me=True, struct_as_record=False)
#dbs = mat["dbStruct"]

utm = mat["Cdb"]

print(mat.keys())
print(type(utm))
print(utm.shape)

distanceX = utm[0,0] - utm[0,1]
distanceY = utm[1,0] - utm[1,1]
print(abs(distanceX))
print(abs(distanceY))

distance = math.sqrt(distanceX*distanceX + distanceY*distanceY)
print(distance)

```
